**[Problem 6] Creating a diagram**

```python
plt.pie(df.groupby("Species").size(), labels=df["Species"].replace(to_replace=[0,1,2], value=["Setosa", "Versicolor", "Virginica"]).unique(), autopct="%.1f%%")
plt.show()
```

```python
for i in range(4):
    boxplot = sns.boxplot(x=df['Species'], y=df.iloc[:, i])
    boxplot.set_xticklabels(np.unique(df["Species"].replace(to_replace=[0,1,2], value=["Setosa", "Versicolor", "Virginica"])))
    plt.show()
```

```python
for i in range(4):
    violinplot = sns.violinplot(x=df['Species'], y=df.iloc[:, i])
    violinplot.set_xticklabels(np.unique(df["Species"].replace(to_replace=[0,1,2], value=["Setosa", "Versicolor", "Virginica"])))
    plt.show()
```

Differences and advantages between boxplot and violinplot

In `boxplot`, the middle line is the median, the top and bottom of the box are the first and third quartiles, and the top and bottom of the whiskers are the maximum and minimum values. Outliers are represented by dots. 

The `violin plot` visualizes the density of the distribution of the data, and a box plot is drawn inside. 

Boxplot can see the variability of the data, but violinplot can also see the distribution density of the data.

**[Problem 7] Confirming the relationship between features**

```python
color = {0: 'red', 1: 'blue', 2: 'yellow'}

for j in range(4):
  for k in range(4):
    if j == k:
      pass
    else:
      plt.xlabel(df.columns[k])
      plt.ylabel(df.columns[j])
      for i in range(len(df)):
        plt.scatter(df.iloc[i, k], df.iloc[i, j], c=color[df.iloc[i, 4]])
      plt.show()
```

It is important to consider when exploring the data. While looking at the scatter plot procession, briefly think about how the three types of flowers are likely to be divided.

The length and width of the petals of setosa are smaller than those of versicolor and virginica. In addition, the length and width of the petals of versicolor and virginica overlap in some parts, but versicolor tends to be smaller. Therefore, it seems possible to classify by petal length and width.

The length of the calyx tends to be smaller in the order of setosa, versicolor, and virginica, but there are many overlapping parts. Also, the calyx lengths of setosa are relatively uniform compared to the other two.

The calyxes are all similar in size and may be difficult to separate.
